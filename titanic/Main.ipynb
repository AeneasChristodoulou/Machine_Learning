```python
import pandas as pd
import numpy as np
import seaborn as sns
df = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
def drop_and_convert(df):
    df.drop(["Name", "Ticket", "Cabin", "PassengerId"], axis = 1, inplace = True)
    df.replace(to_replace = "male", value = 0, inplace = True)
    df.replace(to_replace = "female", value = 1, inplace = True)
    df.replace(to_replace = "S", value = 0, inplace = True)
    df.replace(to_replace = "Q", value = 1, inplace = True)
    df.replace(to_replace = "C", value = 2, inplace = True)
    df.fillna(df.mean(), inplace = True)
    return df

drop_test = drop_and_convert(test)
drop_df = drop_and_convert(df)
```
